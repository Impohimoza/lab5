```python
BATCH_SIZE = 8
EPOCH = 30
VAL_SPLIT = 0.2
CLASS_NUM = 10 
IMG_SIZE = 250
IMG_CHANNELS = 3
input_shape = (IMG_SIZE, IMG_SIZE, IMG_CHANNELS)
RANDOM_SEED = 12
np.random.seed(RANDOM_SEED)
```

<h2>Подготовка данных</h2>

<h3>Data Augmentation</h3>

```python
train_gen = ImageDataGenerator(
    rescale=1. / 255,
    rotation_range=30, #Поворот
    width_shift_range=0.1, #Смещение по ширине
    height_shift_range=0.1, #Смещение по высоте
    brightness_range=[0.5, 1.5], #Изменение яркости
    shear_range=0.2, #Сдвиг изображений
    zoom_range=[0.75,1.25], #Масштабирование изображения
    horizontal_flip=True, #отзеркаливать изображение
    validation_split=VAL_SPLIT,
    
) 
```

```python
train_generator = train_gen.flow_from_directory(
    directory=DATA_PATH+'train',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    seed=RANDOM_SEED,
    shuffle=True,
    subset='training', 
)

valid_generator = train_gen.flow_from_directory(
    directory=DATA_PATH+'train',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    seed=RANDOM_SEED,
    shuffle=True,
    subset='validation', 
)
```

```python
x,y = next(train_generator)
plt.figure(figsize=(12,8))

for i in range(0,6):
    image = x[i]
    plt.subplot(3,3, i+1)
    plt.imshow(image)
    print(y[i])
plt.show()
```

<h2>Построение  модели

```python
base_model = tf.keras.applications.EfficientNetB3(
    weights='imagenet',
    include_top=False,
    input_shape=input_shape,
)
```

```python
base_model.summary()
```

```python
base_model.trainable = False
```

```python
model = models.Sequential()
model.add(base_model)
model.add(layers.GlobalAveragePooling2D(),)
model.add(layers.Dense(units=256, activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.25))

model.add(layers.Dense(CLASS_NUM, activation='softmax'))
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import LearningRateScheduler, ModelCheckpoint, EarlyStopping
from tensorflow.keras.callbacks import Callback
```

<h2>Обучение головы</h2>

```python
for layer in model.layers:
    print(layer, layer.trainable)
```

```python
model.compile(loss="categorical_crossentropy", optimizer=optimizers.Adam(learning_rate=0.001), metrics=["accuracy"])
```

```python
checkpoint = ModelCheckpoint(MODEL_PATH+'step1-{epoch:02d}-{val_loss:.4f}.keras' , monitor = ['val_accuracy'] , verbose = 1, mode = 'max')
earlystop = EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True)
callbacks_list = [checkpoint, earlystop]
```

```python
scores = model.evaluate(valid_generator, verbose=1)
print("Accuracy: %.2f%%" % (scores[1]*100))
```

```python
history = model.fit(
                    train_generator,
                    steps_per_epoch=train_generator.samples//train_generator.batch_size,
                    validation_data=valid_generator, 
                    validation_steps=valid_generator.samples//valid_generator.batch_size,
                    epochs=30,
                    callbacks = callbacks_list
                    )
```
